```python
topjobs=salaries['job_title'].value_counts().sort_values(ascending=False).head(10).reset_index()
topjobs             #topjobs=salaries['job_title'].value_counts().nlargest(10).reset_index()
```

```python
topjobs=pd.DataFrame(topjobs)
topjobs
```

```python
plt.figure(figsize=(10,5))
s1=sns.barplot(topjobs,x='job_title',y='count',color='green')
plt.xticks(rotation=45,ha='right')
plt.title('Top Jobs In IT')
plt.tight_layout()                                    #Top 10 most common job titles
plt.show()
```

Analysis:The above plot shows the top IT jobs, Data Engineering gains more popularity and has a count of 1040 acorss different countries.

```python
#Top 10 company location with most employees
comploc=salaries['company_location'].value_counts().sort_values(ascending=False).reset_index().head(10)
comploc
```

```python
comploc=pd.DataFrame(comploc)
comploc
```

```python
s2=sns.barplot(comploc,x='company_location',y='count',color='green')
plt.title('Top 10 location with most employees')
plt.xlabel('Location')
plt.ylabel('Employees')
plt.tight_layout()                       #Top 10 company location with most employees
plt.show()
```

```python
#Average salary over the years
avgSalaries=salaries.groupby('work_year')['salary'].mean().reset_index().round()
avgSalaries
```

```python
avgSalaries=pd.DataFrame(avgSalaries)
avgSalaries
```

```python
s3=sns.barplot(avgSalaries,x='work_year',y='salary',color='green')
plt.xlabel('Work Year')
plt.title('Salary over the Years')
plt.tight_layout()                           #Average salary over the years
plt.show()
```

```python
#Average salary for the top 10 most common jobs
salary10jobs=salaries.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).reset_index().round().head(10)
salary10jobs
```

```python
salary10jobs=pd.DataFrame(salary10jobs)
salary10jobs
```
